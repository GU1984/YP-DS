```python
# Эффективность обогащения рассчитывается по формуле C * (F - T) / (F * (C - T)) * 100

# C — доля золота в концентрате после флотации/очистки;
C = train.loc[~train['rougher.output.recovery'].isna(), 'rougher.output.concentrate_au']
# F — доля золота в сырье/концентрате до флотации/очистки;
F = train.loc[~train['rougher.output.recovery'].isna(), 'rougher.input.feed_au']
# T — доля золота в отвальных хвостах после флотации/очистки.
T = train.loc[~train['rougher.output.recovery'].isna(), 'rougher.output.tail_au']

# Расчет по формуле
recov_calc = C * (F - T) / (F * (C - T)) * 100

#Расчет по данным
recov_train = train.loc[~train['rougher.output.recovery'].isna(), 'rougher.output.recovery']
```

```python
# расчёт MAE 
mae = mean_absolute_error(recov_train, recov_calc)

print("Показатель MAE между рассчитанным значением и тренировочной выборкой: {0:.2E}".format(mae))
```

### Функция для вычисления итоговой sMAPE

```python
# Функция для расчета Эффективности обогащения.
# Эффективность обогащения рассчитывается по формуле C * (F - T) / (F * (C - T)) * 100, где
# C — доля золота в концентрате после флотации/очистки;
# F — доля золота в сырье/концентрате до флотации/очистки;
# T — доля золота в отвальных хвостах после флотации/очистки.

def smape(true, prediction):
    return sum( abs(true - prediction) / ( (abs(true) + abs(prediction)) * 0.5) ) * 100 / len(true)
```

```python
# Функция для расчета Суммарной эффективности обогащения.
# Суммарная эффективность обогащения рассчитывается по формуле 0.25 * sMAPE(rougher) + 0.75 * sMAPE(final)

def final_smape(smape_rougher, smape_final):
    return 0.25 * smape_rougher + 0.75 * smape_final
```
